## Value of property according to the transacition type

```python
grouped_df = df.groupby(['year', 'transaction']).agg(sum_property_value=('Price', 'sum')).reset_index()

fig, ax = plt.subplots(figsize=(12, 8))

transaction_types = grouped_df['transaction'].unique()
for transaction in transaction_types:
    data = grouped_df[grouped_df['transaction'] == transaction]
    ax.plot(data['year'], data['sum_property_value'], marker='o', label=transaction)

ax.set_title('Sum of Property Values by Year and Transaction Type')
ax.set_xlabel('Year')
ax.set_ylabel('Sum of Property Values')
ax.legend(title='Transaction Type')

plt.show()
```

Price of resale property's is higher than new property as the are mostly funished and they are surrounded by overlookings

## Value of propert Year by Year in Each city

```python
grouped_df = df.groupby(['year', 'city']).agg(sum_property_value=('Price', 'sum')).reset_index()

fig, ax = plt.subplots(figsize=(12, 8))

transaction_types = grouped_df['city'].unique()
for transaction in transaction_types:
    data = grouped_df[grouped_df['city'] == transaction]
    ax.plot(data['year'], data['sum_property_value'], marker='o', label=transaction)

ax.set_title('Sum of Property Values by Year and City')
ax.set_xlabel('Year')
ax.set_ylabel('Sum of Property Values')
ax.legend(title='Citys')

plt.show()
```

```python
num_rows = len(df)

# Calculate the number of registered and not registered entries
num_registered = int(num_rows * 0.7563)
num_not_registered = num_rows - num_registered

# Create the Registry column with the specified proportions
registry_values = ['Yes'] * num_registered + ['No'] * num_not_registered
np.random.shuffle(registry_values)  # Shuffle the values to randomize their order

# Add the Registry column to the DataFrame
df['Furniture'] = registry_values
```
